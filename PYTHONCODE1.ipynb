```python
class p_t:
    '''class to implement an iterator of powes of two'''
    def __init__(self,max=0):
        self.max=max
    def __iter__(self): #start from 0
        self.n=0
        return self
    def __next__(self):
        if self.n<=self.max:
            result=2**self.n
            self.n += 1
            return result
        else:
            raise StopIteration

print(p_t.__doc__)
a=p_t(4)
i=iter(a)
print(next(i))
print(next(i))   
print(next(i))   
print(next(i))   
print(next(i))   
#print(next(i))  
```

```python
class method:
    def __init__(self,fruit):
        self.fruit=fruit
    def __iter__(self):
        return self
    def __next__(self):
        self.fruit+=2
        return self.fruit
    def instance(self):
        print(self.fruit)
    def static():
        print('static method')

m=method(5)
it=iter(m)
print(next(it))
print(next(it))
m.instance()
method.static()
```

```python
#creating class and object in python
class myBird:
    def __init__(self):
        print("myBird 1Class constructor is executing....")
    def whatType(self):
        print("I am a Bird....")
    def canSwim(self):
        print("I can swim")

class myParrot:
    #class attribute
    species="Bird"

    #instance attribute
    def __init__(self,name,age):
        print("myParrot 2class constructor is executing.....")
        self.name=name
        self.age=age

    def canSing(self,thissong):
        return "{} can sing {}....".format(self.name,thissong)

#myParrot class inheriting the attributes from the myBird Class
class myPenguin(myBird):
    def __init__(self):
        #call super() function
        super().__init__()
        print("penguin is ready")
    def whoisThis(self):
        print("i am penguin")
    def canRun(self):
        print("I can run faster")

#instantiate the parrot close
mp1=myParrot("myParrot1",10)
mp2=myParrot("myParrot1",20)

#access the class attributes
print("mp1 is a {}".format(mp1.__class__.species))
print("mp2 is also {}".format(mp2.__class__.species))

#access the instance attributes
print("{} is {} year of age".format(mp1.name,mp1.age))
print("{} is {} year of age".format(mp2.name,mp2.age))
print(mp1.canSing("Chirp"))

#accessing the child's attributes(Inheritance)
pg1=myPenguin()
pg1.whoisThis()
pg1.canRun()

#Data Encapsulation 
class personalComputer:
    def __init__(self):
        self.maxComPrice=20000
    def mySell(self):
        print("Selling Price:{}".format(self.maxComPrice))
    def setmaxComPrice(self,price):
        self.maxComPrice=price

pc=personalComputer()
pc.mySell()
#change the price
pc.maxComPrice=30000
pc.mySell()
#using setter function
pc.setmaxComPrice(40000)
pc.mySell()
        
```

```python
#operator overloading
#this used when the object comes into operation like adding,subtracting,etc
#real example are shopping amzon,flipkart..and other
#in cart we'll add number of items "cart 1+cart 2"
#it helps working of custom objects to meaningful 

class myPoint:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def __str__(self):
        return"{0},{1}".format(self.x,self.y)
    #overloading + and < operation
    def __add__(self,other):
        x=self.x+other.x
        y=self.y+other.y
        return myPoint(x,y)
        
    def __lt__(self,other):
        self_mag=(self.x**2)+(self.y**2)
        other_mag=(other.x**2)+(other.y**2)
        return self_mag<other_mag

p1=myPoint(111,222)
p2=myPoint(14,25)
print(p1)
print(p2)
print()
print(p1<p2)
print(p1+p2)
print()
print(p1.__lt__(p2))
print(p1.__add__(p2))
```

```python
#A generator does not execute completely when called. 
#It returns a generator object (iterator) and executes only when requested.
def test():
    print("Start")
    yield 1
    print("Middle")
    yield 2

g = test()      # Function body NOT executed yet
print(g)        # Generator object

print(next(g))  # Executes until first yield
#print(next(g))  # Continues from previous position
```

```python
#generators with a loop
def reverse_string(my_string):
    length=len(my_string)
    for i in range(length-1,-1,-1):#(length func,step minus 1,upto 0 it iterate)
        yield my_string[i]
for char in reverse_string("PYTHON"):
    print(char)
```

```python
#Decorators
#A decorator is a function that takes another function as an argument, 
#adds extra functionality to it without modifying its original code, 
#and returns the modified function.
def make_decorated(func):
    def inner_func():
        print("I got decorated")
        func()
    return inner_func()

def simple_func():
    print("I am a simple function")

decor=make_decorated(simple_func)
decor()
```

```python
#Another code from demonstration
def my_div(func):
    def inner_func(x,y):
        print("i am dividing",x,"and",y)
        if y==0:
            print("oops!this number divides by 0")
            return
        return func(x,y)
    return inner_func

@my_div
def go(a,b):
    return a/b

a=int(input("enter a value"))
b=int(input("enter a value"))
go(a,b)

#or else we can gives the parameter on function calling itself
#print(go(10,5))
```
